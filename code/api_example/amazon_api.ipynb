```python
import requests
import json
import pandas as pd

pd.set_option('display.max_colwidth', None)
pd.set_option('display.max_rows', None)
```

```python
f = open("rf_api.txt", "r")
api_key = f.read()
```

```python
params = {
  'api_key': api_key,
  'type': 'category',
  'amazon_domain': 'amazon.com',
  'category_id': '11060451' # cid for skin care
}

api_result = requests.get('https://api.rainforestapi.com/request', params)

api_response = api_result.json()
```

```python
names = []
links = []
for i, key in enumerate(api_response['refinements'].keys()):
    names.extend([x['name'] for x in api_response['refinements'][key]])
    links.extend([x['link'] for x in api_response['refinements'][key]])
```

```python
link_df = pd.DataFrame(names, columns=['category'])
link_df['link'] = links
```

```python
params = {
  'api_key': api_key,
  'type': 'bestsellers',
  'url': 'https://www.amazon.com/Best-Sellers-Beauty-Body-Moisturizers/zgbs/beauty/11060661/ref=zg_bs_nav_bt_3_11060521'
}

api_result = requests.get('https://api.rainforestapi.com/request', params)

print(json.dumps(api_result.json()))
```

```python
aid = [x['asin'] for x in api_result.json()['bestsellers']]
```

```python
for a in aid[:5]: 
    params = {
      'api_key': api_key,
      'type': 'product',
      'amazon_domain': 'amazon.com',
      'asin': a
    }

    api_result = requests.get('https://api.rainforestapi.com/request', params)

    res_dict = api_result.json()
    
    with open(f'{a}.json', 'w') as f:
        json.dump(res_dict, f)
```
