### Load the api key

```python
f = open("rf_api.txt", "r")
api_key = f.read()
f.close()
```

### Step 1
A user opens our website and searches a product. 
<br>
<br>
Assumption:
- the product belongs to skin care category.

The search term can be anything, including brand / subcategory / brand+subcategory, eg. lancome / night cream / lancome night cream
<br>
<br>
Note: Every api search can only return one page of results, if more results are required, we need to change the parameter 'page = 2', and products on page 2 can be pulled. Every page would cost one api search. 

```python
params = {
  'api_key': api_key,
  'type': 'search',
  'search_term': 'fresh', # a brand
  'amazon_domain': 'amazon.com',
#   'page': '1',
  'category_id': '11060451' # cid for skin care
}

api_result = requests.get('https://api.rainforestapi.com/request', params)

api_response = api_result.json()
```

How many products does the search return?
<br>
It may vary with different search terms. Usually 20-60 products will be returned on the first page. 
<br>

```python
len(api_response['search_results'])
```

What can we know each product pulled from the search results?

```python
# Some features for product 1 (index 0) might not be available for other products, such as price
api_response['search_results'][0].keys()
```

```python
# The second product in the picture does not have price.
Image(filename='noprice.png') 
```

On the website I suggest we can display the following information which can help users find the target product
- title
- image
- price

Hopefully the target product that the user want to find is within the above page 1 list.  

Note: filter out the sponsored products; they are usually not closely related to the search.

```python
titles = [x['title'] for x in api_response['search_results']]
images = [x['image'] for x in api_response['search_results']]
sponsored_flags = [x['sponsored'] for x in api_response['search_results']]
asins = [x['asin'] for x in api_response['search_results']]
prices = [x['price']['value'] if 'price' in x.keys() else None for x in api_response['search_results']]


info_list = pd.DataFrame(titles, columns=['title'])
info_list['asin'] = asins
info_list['price'] = prices
info_list['image'] = images
info_list['sponsored'] = sponsored_flags

# for a clear look we probably only display title, price and image for sponsored = False
# asin is for the next step but not displayed

info_list
```

### Step 2
If the user clicks one product among the list, then we need to use api again to retrieve important information NOT included in the above search result
- description for users to check the product's effect
- ingredient for the EWG analysis (aims to provide the product's potential risks) 

The given information is the asin (product id)

```python
params = {
      'api_key': api_key,
      'type': 'product',
      'amazon_domain': 'amazon.com',
      'asin': 'B00AU9EBTC'
}

api_result = requests.get('https://api.rainforestapi.com/request', params)

api_response = api_result.json()
```

The api search gives a lot information, but we mainly focus on 
- description
- ingredients

```python
api_response['product'].keys()
```

The description gives users an overview (mainly benefits) about the product.

```python
api_response["product"]['description']
```

The ingredients might tell users some potential risks of the product.
<br>
We put the extracted ingredients into the ml model generated from ewg data.

#### Important note:
not every skin care product on amazon has the ingredients explicitly listed out
- some do not have ingredient section at all
- some might list in the description 

However, the formal way for a seller to list all ingredients is using the below section 

```python
sections = api_response["product"]["important_information"]['sections']
for section in sections:
    if section['title'] == 'Ingredients':
        ingredients = section['body']
ingredients.strip(',').split(', ')
```

Meanwhile we also need to know the category in case the product is not good enough and needs replacements.

```python
categories = [x['name'] for x in api_response["product"]["categories"]]
# the last one is the most specific
categories
```

### Step 3
use the api the third time to get the best seller products in the subcategory

They are our candidates for replacements, and we need to further analyze them in the step 4 for the final recommendation.

tricky part: the subcategory urls (I'll explain how to get them in the meeting - Sophie)

Note: the results can be saved, though the best-seller rating is dynamic, we can update the top 50 products for face cream / face washes / eye serums and etc every week.

```python
params = {
  'api_key': api_key,
  'type': 'bestsellers',
  'url': 'https://www.amazon.com/Best-Sellers-Beauty-Facial-Cleansing-Washes/zgbs/beauty/7730193011/ref=zg_bs_nav_bt_4_11060901'
}

api_result = requests.get('https://api.rainforestapi.com/request', params)
```

```python
aid = [x['asin'] for x in api_result.json()['bestsellers']]
aid
```

### Step 4
for each product id we need to pull the description, ingredients information to analyze in our ml model and then get the top ones for the final recommendation. so in this step, the api search count is like 10-50. However, the good news is that if we only update the best-seller info every weeks, the information can be saved for each sub category. 
<br>
The approximation for total api search count is 30 (top 30 best-seller products) * 30 subcategory = 900 api search


An example for saving 5 products infomation.

```python
for a in aid[:5]: 
    params = {
      'api_key': api_key,
      'type': 'product',
      'amazon_domain': 'amazon.com',
      'asin': a
    }

    api_result = requests.get('https://api.rainforestapi.com/request', params)

    res_dict = api_result.json()
    
    with open(f'{a}.json', 'w') as f:
        json.dump(res_dict, f)
```
