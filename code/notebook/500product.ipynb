```python
def extract_from_rainforest(result: dict) -> dict:
    """
    Extracts all needed information from Rainforest API request in preparation to store in cached_data.amazon_products
    database.
    Lists datatypes converted to patterned strings to avoid headache of datatype declaration in RDS database
    Some of these may need to be tested further (i.e. ingredients, skin type, brand) since their location seemed to 
    vary for each request
    :return: dictionary with keys being the schema of cached_data.amazon_products 
    """
    target_product = dict()
    info = result['product']
    
    # asin
    asin = info['asin']
    target_product['asin'] = asin
    
    # product link
    product_link = info['link']
    target_product['product_link'] = product_link
    
    # product name
    product_title = info['title']
    target_product['product_title'] = product_title
    
    # product category
    product_category = info['categories'][-1]['name']
    target_product['product_category'] = product_category
    
    #product key words
    product_keywords = result['product']['keywords']
    target_product['product_keywords'] = product_keywords

    # product description
    target_product['product_description'] = 'Not Available'
    if 'description' in info.keys():
        target_product['product_description'] = info['description']
    elif 'a_plus_content' in info.keys():
        if 'company_description_text' in info['a_plus_content'].keys():
            target_product['product_description'] = info['a_plus_content']['company_description_text']
      
    # feature_bullets   
    target_product['feature_bullets'] = 'Not Available'
    if 'feature_bullets' in info.keys():
        target_product['feature_bullets'] = info['feature_bullets_flat']
        
    #image_url
    image_url = info['main_image']['link']
    target_product['image_url'] = image_url
    
    # ingredients
    target_product['ingredients'] = 'Not Available'
    if 'important_information' in info.keys():
        for important_info in info['important_information']['sections']:
            if 'title' in important_info.keys() and important_info['title'] == 'Ingredients':
                ingredients = important_info['body']
                target_product['ingredients'] = ingredients
                
    # pulled from "attributes" subdirectory
    skin_type = "Not Available"
    brand = "Not Available"
    if 'attributes' in info.keys():
        for attr in info['attributes']:
            if attr['name'] == 'Skin Type':
                skin_type = attr['value']
            if attr['name'] == 'Brand':
                brand = attr['value']
    target_product['skin_type'] = skin_type
    target_product['brand'] = brand

    return target_product
```

```python
ingredient = []
title = []
asin = []
category = []
brand = []
skin_type = []
product_link = []
image_url = []  
description = []
feature_bullets = []

directory = './face_moisturizer'
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        with open(os.path.join(directory, filename)) as f:
            result = json.load(f)
        target = extract_from_rainforest(result)
        ingredient.append(target['ingredients'])
        title.append(target['product_title'])
        category.append(target['product_category'])
        asin.append(target['asin'])
        brand.append(target['brand'])
        skin_type.append(target['skin_type'])
        product_link.append(target['product_link'])
        image_url.append(target['image_url'])
        description.append(target['product_description'])
        feature_bullets.append(target['feature_bullets'])
    else:
        continue
```

```python
for p in range(1,23):
    params = {
    'api_key': api_key,
    'type': 'category',
    'amazon_domain': 'amazon.com',
    'category_id': '16479981011',
    'page': p
    }
    
    api_result = requests.get('https://api.rainforestapi.com/request', params)
    json_result = api_result.json()
        
    # store api results as jsons in api_result directory
    with open(f'./api_result/page_{p}.json', 'w') as f:
        json.dump(json_result, f)

print('Done!')
```

```python
# Read in saved api_results jsons to save asin and prices as df
api_result_asin = []
api_result_prices = []


directory = './api_result'
for filename in os.listdir(directory):
    if filename.endswith(".json"):
        with open(os.path.join(directory, filename)) as f:
            result = json.load(f)
        product_list = result['category_results']
        for product in product_list:
            api_result_asin.append(product['asin'])
            if 'prices' in product.keys():
                api_result_prices.append(product['prices'][0]['value'])
            else:
                api_result_prices.append('')
#         target = extract_from_rainforest(result)
#         target = pd.DataFrame.from_records(result.json()['category_results'])
#         asin.append(result['category_results']['asin'])
#         prices.append(target['feature_bullets'])
    else:
        continue
print('Done!')
```
