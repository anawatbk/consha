# Interacting with Consha Database

## Connect to Database:

```python
import sqlalchemy as db
```

```python
engine = db.create_engine(Config.SQLALCHEMY_DATABASE_URI)
conn = engine.connect()
metadata = db.MetaData()
```

## Browse MetaData:

```python
insp = db.inspect(engine)
insp.get_schema_names()

schema='aws_commons'

for i in insp.get_schema_names():
    print(f"Schema: {i}")
    tables = insp.get_table_names(schema=i)
    print(f"Tables: {tables}")
    print()
```

## Querying from Database:

[Querying Use Cases (Towards Data Science)](https://towardsdatascience.com/sqlalchemy-python-tutorial-79a577141a91)

```python
metadata = db.MetaData(schema='crawled_data')
ewg = db.Table('ewg_product', metadata, autoload=True, autoload_with=engine)
ewg.columns.keys()
```

```python
query = db.select([ewg.columns.id, ewg.columns.product_name, ewg.columns.ingredient, ewg.columns.product_score])
result = conn.execute(query).fetchall()
result
```
